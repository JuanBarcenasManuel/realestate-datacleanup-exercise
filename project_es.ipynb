#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
ds = pd.read_csv('https://raw.githubusercontent.com/JuanBarcenasManuel/realestate-datacleanup-exercise/refs/heads/main/assets/real_estate.csv', sep=';')

ds['pps'] = ds['price'] / ds['surface']
valdemorillo_pps = ds[ds['level5'] == 'Valdemorillo']['pps']
galapagar_pps = ds[ds['level5'] == 'Galapagar']['pps']

mean_pps_valdemorillo = valdemorillo_pps.mean()
mean_pps_galapagar = galapagar_pps.mean()

print(f"El promedio de precio por metro cuadrado en Valdemorillo es: {mean_pps_valdemorillo:.2f} EUR/m2")
print(f"El promedio de precio por metro cuadrado en Galapagar es: {mean_pps_galapagar:.2f} EUR/m2")
print("En conclusion, el precio en Valdemorillo es mas economico")
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

```python
import pandas as pd

ds = pd.read_csv('https://raw.githubusercontent.com/JuanBarcenasManuel/realestate-datacleanup-exercise/refs/heads/main/assets/real_estate.csv', sep=';')

southern_belt_populations = ['Getafe', 'Alcorcón']
ds_southern_belt = ds[ds['level5'].isin(southern_belt_populations)]
ds_southern_belt['pps'] = ds_southern_belt['price'] / ds_southern_belt['surface']
mean_pps_getafe = ds_southern_belt[ds_southern_belt['level5'] == 'Getafe']['pps'].mean()
mean_pps_alcorcon = ds_southern_belt[ds_southern_belt['level5'] == 'Alcorcón']['pps'].mean()
print(f"El precio por metro cuadrado medio en Getafe es: {mean_pps_getafe:.2f} EUR/m2")
print(f"El precio por metro cuadrado medio en Alcorcón es: {mean_pps_alcorcon:.2f} EUR/m2")
print("El precio de ALcaron es mas economico")
```
